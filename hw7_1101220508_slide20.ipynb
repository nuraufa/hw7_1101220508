```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
# Mengambil data CSV
data = np.loadtxt('1.csv', delimiter=',', usecols=(1, 2, 3, 4), dtype='int32')

# Mengambil subset data (1000-1100)
y = data[1000:1100, 1]

# Fungsi untuk mencari puncak dengan jarak minimal tertentu
def find_peaks_custom(data, min_distance):
    peaks = []
    for i in range(1, len(data) - 1):
        if data[i - 1] < data[i] > data[i + 1]:  # Puncak
            if not peaks or i - peaks[-1] >= min_distance:  # Jarak antar puncak
                peaks.append(i)
    return peaks

# Mencari puncak dengan jarak minimal 10 dan 3
peaks_d10 = find_peaks_custom(y, 10)
peaks_d3 = find_peaks_custom(y, 3)

# Mencetak indeks puncak yang ditemukan
print('distance 10 idx:', peaks_d10)
print('distance 3 idx:', peaks_d3)

# Menampilkan grafik
plt.figure(figsize=(12, 5))
plt.plot(y, label='Data')
plt.scatter(peaks_d3, y[peaks_d3], marker='v', color='r', label='distance 3')
plt.scatter(peaks_d10, y[peaks_d10], marker='x', color='k', label='distance 10')
plt.legend()
plt.title('Puncak dengan Distance 3 dan 10')
plt.xlabel('Indeks')
plt.ylabel('Nilai')
plt.grid(True)
plt.show()
```
