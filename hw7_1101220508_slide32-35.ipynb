```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pathlib
from sklearn.ensemble import RandomForestClassifier
import os
```

```python
# Pilih 100 sampel pelatihan
n = 100
np.random.seed(100)

# Daftar file dataset
files = ['1.csv', '2.csv', '3.csv', '4.csv', '5.csv', '6.csv', '7.csv', '8.csv']

# List untuk menyimpan data
x_data_list = []
y_data_list = []

# Proses setiap file dataset
for idx, file in enumerate(files):
    print(f"Memproses {file}")
    
    # Membaca dataset dari file CSV
    df = pd.read_csv(file)
    
    # Mengisi nilai hilang dengan forward fill dan backward fill
    df.ffill(inplace=True)  # Forward fill
    df.bfill(inplace=True)  # Backward fill
    
    # Mengonversi DataFrame menjadi numpy array
    data_array = df.values
    
    # Pilih n sampel secara acak
    selected_idx = np.random.choice(data_array.shape[0], min(n, data_array.shape[0]), replace=False)
    selected_data = data_array[selected_idx]
    
    # Menambahkan data dan label
    x_data_list.append(selected_data)
    y_data_list.append(np.full(selected_data.shape[0], idx))
    
    print(f'Shape data: {df.shape}, memilih {selected_data.shape[0]} sampel')

# Menggabungkan seluruh data dan label
x_data_combined = np.vstack(x_data_list)
y_labels_combined = np.hstack(y_data_list)

# Menampilkan informasi tentang data
print(f'Jumlah data: {x_data_combined.shape[0]}')
print(f'Jumlah fitur: {x_data_combined.shape[1]}')
print(f'Jumlah label: {y_labels_combined.shape[0]}')

# Inisialisasi RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)

# Melatih model dengan data
rf_classifier.fit(x_data_combined, y_labels_combined)

# Menghitung akurasi pada data pelatihan
accuracy = rf_classifier.score(x_data_combined, y_labels_combined)
print(f'Akurasi Model: {accuracy}')

# Mengambil pentingnya fitur
feature_importances = rf_classifier.feature_importances_

# Menyusun fitur dan pentingnya
feature_names = list(df.columns)  # Mengambil nama fitur dari dataset terakhir
sorted_feature_importance = sorted(zip(feature_importances, feature_names), reverse=True)

# Menampilkan urutan pentingnya fitur
print('Pentingnya Fitur yang Telah Diurutkan:')
for importance, feature in sorted_feature_importance:
    print(f'{feature}: {importance}')

# Menyimpan hasil pentingnya fitur ke dalam file CSV
output_dir = r'D:\Output HW 7 Slide 32-35 new'

# Pastikan folder tujuan ada
os.makedirs(output_dir, exist_ok=True)

# Menyimpan hasil fitur penting
for idx, (importance, feature) in enumerate(sorted_feature_importance):
    output_file_path = os.path.join(output_dir, f'feature_importance_{idx}.csv')
    with open(output_file_path, 'w') as f:
        f.write("Feature, Importance\n")
        f.write(f"{feature}, {importance}\n")

print(f"Hasil fitur penting disimpan di {output_dir}")
```
